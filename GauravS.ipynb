```python
product_extension = "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"
product_detail_span_class = "a-list-item"
```

```python
time.sleep(5)

product_list = driver.find_elements(By.XPATH, f"//a[@class='{product_extension}']")

product_hrefs = [webelem.get_attribute('href') for webelem in product_list]
print(len(product_hrefs))
```

```python
def scrape_page(link):
    #Get to the specified link
    try:
        driver.get(f"{link}")
    except:
        return
    #click on see all reviews
    seeAllReviews = driver.find_element(By.CSS_SELECTOR, ".a-link-emphasis.a-text-bold")
    driver.get(seeAllReviews.get_attribute('href'))
   

    prodTitle = driver.current_url[driver.current_url.index("product-reviews/") + 16:driver.current_url.index("product-reviews/") + 26]
    review_dict = []
    index = driver.current_url.index("all_reviews")
    driver.get(driver.current_url[0:index] + "avp_only_reviews" + driver.current_url[index + 12:])

    
    #getting text from all reviews on the page
    reviewElem = driver.find_elements(By.CSS_SELECTOR, ".a-section.review.aok-relative")
    complete = True
    
    #algorithm to click on next page
    while (complete):

        for webelem in reviewElem:
            review = []
            try:
                startext = webelem.find_element(By.CSS_SELECTOR, "div>div>.a-row>.a-link-normal>[data-hook=review-star-rating]>.a-icon-alt") #Review text
                reviewtext = webelem.find_element(By.CSS_SELECTOR, ".a-size-base.review-text.review-text-content")
                review.append(prodTitle)
                review.append(reviewtext.text)
                review.append(startext.get_attribute('innerHTML'))
                review_dict.append(review)
            except:
                print("Error")
        #going to next page
        try:
            #getting and clicking href of next page
            nextPage = driver.find_element(By.CSS_SELECTOR, ".a-last > a")
            driver.get(nextPage.get_attribute('href'))
            reviewElem = driver.find_elements(By.CSS_SELECTOR, ".a-section.review.aok-relative")
            print('next page')
        except:
            print('no more pages')
            complete = False
    print('done')
    print(len(review_dict))
    return review_dict
```

```python
product_list = []
#change this value to increase number of products
num_of_products = 2
for product in product_hrefs[:num_of_products]:
    product_list += scrape_page(product)

driver.close()
```

```python
import csv
  
with open('data1.csv', 'w') as f:
    csv_writer = csv.writer(f)
    for product in product_list:
        csv_writer.writerow(product)
```
